```python
class Weapon:
    def setDamage(self, damage):
        self.damage = damage

class WeaponEnhancement(Weapon):
    def setEnhancement(self):
        import random
        percent = random.randint(0, 100)
        if percent >= 40:
            self.damage *= 2
            self.log = '성공'
            return '강화성공!'
        else:
            if self.damage <= 1:
                self.damage = 2
            self.damage /= 2
            self.log = '실패'
            return '강화실패!'
class Equipment(WeaponEnhancement):
    pass

class Role:
     def setRole(self, role):
        self.role = role

class Skill:
    def slash(self):
        return '베기!'
    def beam(self):
        return '빔!!'

class Body(Role, Skill, Equipment):
    def __init__(self, role, hp, mp, power):
        self.setRole(role)
        self.hp = hp
        self.mp = mp
        self.power = power


hero = Body('hero', 100, 50, 50)
hero.setDamage(500)
print(hero.slash())
print(hero.damage)
print(hero.beam())
hero.setEnhancement()
print(hero.damage)
```

```python
#데미지 초기화 
hero.damage = 1
```

```python
count = 1
myscore = []
mylog = []
while True:
    if hero.damage >= 100000:
        break
    print(hero.setEnhancement(), f'{count}번째 시도입니다.')
    count += 1
    print(hero.damage)
    myscore.append(int(hero.damage))
    mylog.append(hero.log)
```

```python
from collections import Counter
print(sorted(myscore, reverse = True)[:30])
Counter(mylog)

rate = Counter(mylog)['성공'] / (Counter(mylog)['성공'] + Counter(mylog)['실패'])
rate
```
