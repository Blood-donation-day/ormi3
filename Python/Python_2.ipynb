##비교 연산자

```python
#기존 내용과 굉장히 유사함으로 짧게 필수 예제만 하고 넘어감

print(5 == 5) # 결과: True

print(5 != 3) # 결과: True

print(3 < 5) # 결과: True
print(5 <= 5) # 결과: True

print(6 > 5)  # 결과: True
print(5 >= 5) # 결과: True

print('apple' < 'banana') # 결과: True

#위처럼 알파벳은 구분하기가 쉽지만 한글은 쉽지 않습니다.
#유니코드표에서 하는 D558번째에 있으며 후는 D6C4번째에 있습니다.
print('하하' > '후후') # 결과: False

x = 10
print(5 < x < 15) # 결과: True
```

```python
#리스트의 비교는 맨 앞자리를 비교합니다. (같다면 다음 인덱스로 넘어가서 비교)
[10, 200, 30] > [20, 30, 40] # False
```

```python
#비교연산 산술연산은 모두 class에서 구현이 가능합니다.
#비교 예제

class int(int):
    def __eq__(self, next):
        return True

# 10 > 11
int('10') == int('11')
```

```python
#더하기 예제
class int(int):
    def __add__(self, next):
        return self * next

# 10 + 11
int(10) + int(11)
```
