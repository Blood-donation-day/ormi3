##클래스 변수와 인스턴스 변수

```python
class Car:
    # 클래스 변수의 위치
    max_speed = 300
    max_people = 5
    car_gear = ['P', 'D', 'R', 'N']

modelx = Car()
modely = Car()

modely.max_speed = 500  # 인스턴스 변수를 변경 (클래스 변수 X)

print(id(Car.max_speed))    #300
print(id(modelx.max_speed)) #300
print(id(modely.max_speed)) #500(인스턴스 변수를 출력)
```

```python
class Car:
    kinds = []
    speed = 300

    def add_kinds(self, name):  #kinds 라는 list가 변수에 없음으로 class변수로 들어감
        self.kinds.append(name)

    def change_speed(self, speed): #인스턴스 변수
        self.speed = speed

modelx = Car()
modely = Car()

modelx.add_kinds('x')
modely.add_kinds('y')

Car.speed = 100

modelx.change_speed(500)
modely.change_speed(250)

print(f'modelx.kinds: {modelx.kinds}') #add_kinds안에 kind라는 list가 없음으로 class의 변수를 가져옴
print(f'modely.kinds: {modely.kinds}')
print(f'modelx.speed: {modelx.speed}')
print(f'modely.speed: {modely.speed}')
```

##다중상속

```python
#다중상속

class A:
    def hello(self):
        print('hello A')

class B:
    def hello(self):
        print('hello B')

class C(A, B):
    pass

c = C()
c.hello() #주의! python의 코드의 특성상 hello B라고 오해할만 합니다. 2.x, 3.x 상속순서가 다릅니다.
```

```python
#주의! python의 코드의 특성상 hello B라고 오해할만 합니다. 2.x, 3.x 상속순서가 다릅니다.

class A:
    def hello(self):
        print('hello A')

class B(A):
    def hello(self):
        print('hello B')

class C(A):
    def hello(self):
        print('hello C')

#다이아몬드 상속
class D(B, C):
        pass

c = C()
c.hello()

d = D()
d.hello()
print(D.mro()) #Class D B C A object
```

```python
#다중상속에서 에러가 발생하는 경우
#상속 순서에 맞지않게 불러 들렸을 때 예시

class A:
    def hello(self):
        print('hello A')

class B(A):
    def hello(self):
        print('hello B')

class C(B):
    def hello(self):
        print('hello C')

#다이아몬드 상속
class D(B, C):
        pass

d = D()
d.hello()
```

```python
class Weapon:

    def setWeapon(self, weapon):
        self.weapon = weapon

    def setDamage(self, damage):
        self.damage = damage

class WeaponEnhancement(Weapon):
    def setEnhancement(self):
        import random
        percent = random.randint(1, 100)
        if percent >= 50:
            self.damage *= 2
            self.log = '성공'
            return '강화성공!'
        else:
            if self.damage <= 100:
                self.damage = 100
            self.damage /= 2
            self.log = '실패'
            return '강화실패!'
class Equipment(WeaponEnhancement):

    def setquipment(self, helmet =' ', top = '', bottom = '', earring = ''):
        self.helmet = helmet
        self.top = top
        self.bottom = bottom
        self.earring = earring

class Role:
     def setRole(self, role):
        self.role = role

class Skill:
    def slash(self):
        return '베기!'
    def beam(self):
        return '빔!!'

class Body(Role, Skill, Equipment):
    def __init__(self, role, hp, mp, power, weapon):
        self.setRole(role)
        self.hp = hp
        self.mp = mp
        self.power = power
        self.setWeapon(weapon)

hero = Body('hero', 100, 50, 50, '용사의 검')
hero.setquipment('가죽헬멧', '속옷', '', '귀걸이')
print(hero.beam())
print(hero.earring)
```
